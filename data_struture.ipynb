#### 1: List

#### Q1:-Print positive and negative elements of an List

```python
list = [-1,2,-2,3,-4,5,6,-6]
neg = []
post = []

for i in list:
    if i >= 0:
        post.append(i)
    else:
        neg.append(i)
print(f"negative number list is:-{neg}\npositive number list is:-{post}")
```

#### Q2:- Mean of List elements


```python
list = [1,2,3,4,5,6,7,8,9,81]
sum = 0

for i in list:
    sum += i
print(f"mean of list is:- {sum / len(list)}")
```

#### Q3:- Find the greatest element and print its index too

```python
list = [1,2,3,4,5,6,7,8,9,81]
great = 0
index = 0

for i in range(0,len(list),1):
    if list[i] > great:
        great = list[i]
        index = i
        
print(f"greater number from list is:- {great} and index is:- {index}")
```

#### Q4:- Find the second greatest element.

```python
list = [1,2,3,4,5,6,7,8,9,21,34,56,78,98]
large = 0
sec_large = 0

for i in list:
    if i > large:
        large = i
        for i in list:
            if sec_large < i < large:
                sec_large = i
print(f"second largest number is:- {sec_large}")
        
```

#### Q5:- Check if List is sorted or not.

```python
list = [1,2,3,4,5,6,7,8,9]
list_sort = sorted(list)

if list == list_sort:
    print("list is sorted")
else:
    print("list not sorted")
```
