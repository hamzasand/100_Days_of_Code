### For Loop


#### Q1: Accept an integer and Print hello world n times

```python
num = int(input("Please Enter a number"))
for i in range(0,num):
    print("Hello World")
```

#### Q2: Print natural number up to n

```python
num = int(input("Please Enter a number"))
for i in range(0,num):
    print(i)
```

#### Q3: Reverse for loop. Print n to 1

```python
num = int(input("Enter the number"))
for i in range(num,0,-1):
    print(i)
```

#### Q4: Take a number as input and print its table

```python
num = int(input("enter a number for table printing"))

for i in range(1,11,1):
    print(f"{i}x{num} = {i*num}")
```

#### Q5:Sum up to n terms

```python
num = int(input("Enter a number"))
sum = 0
for i in range(1,num,1):
    sum += i
print(sum) 
```

#### Q6: Factorial of a number

```python
num = int(input("Enter a number for factorial"))
fact = 1
for i in range(1,num+1,1):
    fact *= i
print(fact)
```

#### Q7: Print the sum of all even & odd numbers in a range separately

```python
num = int(input("Enter a number to fin odd even sum"))
sum_even = 0
sum_odd = 0

for i in range(1,num+1,1):
    if i % 2 == 0:
        sum_even += i
    else:
        sum_odd += i
print(f" sum of even are:- {sum_even}")
print(f" sum of odd are:- {sum_odd}")
```
