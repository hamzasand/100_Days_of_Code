#### 1: List

#### Q1:-Print positive and negative elements of an List

```python
list = [-1,2,-2,3,-4,5,6,-6]

neg = []

post = []

for i in list:
    if i >= 0:
        post.append(i)
    else:
        neg.append(i)
print(f"negative number list is:-{neg}\npositive number list is:-{post}")
```

#### Q2:- Mean of List elements


```python
list = [1,2,3,4,5,6,7,8,9,81]

sum = 0

for i in list:
    sum += i
print(f"mean of list is:- {sum / len(list)}")
```

#### Q3:- Find the greatest element and print its index too

```python
list = [1,2,3,4,5,6,7,8,9,81]
great = 0
index = 0

for i in range(0,len(list),1):
    if list[i] > great:
        great = list[i]
        index = i
        
print(f"greater number from list is:- {great} and index is:- {index}")
```

### Dictinaries in python

#### Q1:- Write a Python script to merge two Python dictionaries.

```python
dict1 = {"name":"hamza", "age":26, "city":"faisalabad"}
dict2 = {"profession":"Developer", "experience":3}

for i in dict2:
    dict1[i] = dict2[i]

print(dict1)
```

#### Q4:-Write a Python program to sum all the values in a dictionary


```python
dict = {1:10, 2:20, 3:30, 4:40, 5:50}
sum =0

for i in dict:
    sum += dict[i]

print(f"sum of the dictionary is:- {sum}")
```

#### Q5:- Write a Python program to combine two dictionary by adding values for common keys.

```python
d1 = {1:10, 2:20, 3:30, 4:50}
d2 = {4:40, 5:50, 6:60}

for i in d2:
    if i in d1.keys():
        d1[i] =d1[i] + d2[i]
    else:
        d1[i] = d2[i]
print(d1)
```
