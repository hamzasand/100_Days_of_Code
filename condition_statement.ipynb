#### Q1. Accept two numbers and print the greatest between them.

```python
num1 = int(input("Enter your number A:-"))
num2 = int(input("Enter your number B:-"))

if num1 > num2:
    print(f"num1 = {num1} greater then num2={num2}")
elif num1 < num2:
    print(f"num2 = {num2} is greater then num1={num1}")

else:
    print(f"{num2} and {num1} are equal")
```

### 2. Accept the gender from the user as char and print the respective greeting message 
Ex - Good Morning Sir (on the basis of gender)

```python
char = input("enter you gender:-").lower()
if char == "female" or char == "f":
    print('Good Morning Miss')
elif char == "male" or char =="m":
    print("Good Morning Sir")
else:
    print("Undefined gender character")
```

### Q3. Accept an integer and check whether it is an even number or odd.

```python
num = int(input("please enter a number to check odd or even:-"))
if num % 2 == 0:
    print(f"{num} is even")
else:
    print(f"{num} is odd")
```

### Q4. Accept name and age from the user. Check if the user is a valid voter or not.
Ex- “hello shery you are a valid voter”

```python
name = input("Hello, Please enter you name:-")
age = int(input("Now enter your age:-"))

if age >= 18:
    print(f"Hello {name } , you can cast your vote")
else:
    remaining_years = 18 - age
    print(f"Hello {name } , you can not cast your vote, you can cast vote after {remaining_years} years")
```

### 5. Accept a year and check if it a leap year or not (google to find out what is a leap year)

```python
year = int(input("Enter a year to check if it is a leap year: "))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(f"{year} is a leap year")
else:
    print(f"{year} is not a leap year")
```
