### While loop

#### Q1:- Separate each digit of a number and print it on the new line

```python
dig = int(input("Enter a digit"))

while 0<dig:
    print(dig%10)
    dig = dig // 10
```

#### Q2:-Accept a number and print its reverse.

```python
dig = int(input("Enter digit"))
rev_dig = 0
while 0 < dig:
    rev_dig = rev_dig * 10 + dig % 10
    dig = dig // 10
print(rev_dig)
```

#### Q4:-Accept a number and check if it is a pallindromic number (If number and its reverse are equal)

```python
dig = int(input("enter a digit"))
dig1 = dig
rev_dig = 0
while 0 < dig1:
    rev_dig = rev_dig*10 + dig1 %10
    dig1 = dig1//10
if rev_dig == dig:
    print(f"{dig}is a palidram number")
else: 
    print(f"{dig}not palidram") 
```
