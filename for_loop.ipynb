#### Q1: Accept an integer and Print hello world n times

```python
num = int(input("Please Enter a number"))
for i in range(0,num):
    print("Hello World")
```

#### Q3: Reverse for loop. Print n to 1

```python
num = int(input("Enter the number"))
for i in range(num,0,-1):
    print(i)
```

#### Q6: Factorial of a number

```python
num = int(input("Enter a number for factorial"))
fact = 1
for i in range(1,num+1,1):
    fact *= i
print(fact)
```

#### Q7: Print the sum of all even & odd numbers in a range separately

```python
num = int(input("Enter a number to fin odd even sum"))
sum_even = 0
sum_odd = 0

for i in range(1,num+1,1):
    if i % 2 == 0:
        sum_even += i
    else:
        sum_odd += i
print(f" sum of even are:- {sum_even}")
print(f" sum of odd are:- {sum_odd}")
```

#### Q8: Print all the factors of a number

```python
num = int(input("Enter a number to fin odd even sum"))
factors = []
for i in range(1,num+1,1):
    if num % i ==0:
        factors.append(i)
print(factors)
```

#### Q9: Accept a number and check if it a perfect number or not. 
A number whose sum of factors is equal to the number itself 
Ex -  6 = 1, 2, 3 =6

```python
num = int(input("enter a number"))
factors = []
num2 = 0

for i in range(1,num,1):
    if num%i == 0:
        factors.append(i)
        num2 += i
if num2 == num:
    print(f"{num} is a perfect number beacuse sum of factors is {num2}")
else:
    print(f"{num} is a not perfect number beacuse sum of factors is {num2}")

print(factors)
```

#### Q10: Check wether the number is prime or not


```python
number = int(input("Enter a number: "))

if number > 1:
    for i in range(2, number):
        if number % i == 0:
            print(f"{number} is not a prime number")
            break
    else:
        print(f"{number} is a prime number")
else:
    print(f"{number} is not a prime number")
```

#### Q13: Count all letters, digits, and special symbols from a given string 
- Given: str1 = "P@#yn26at^&i5ve" 
- Expected Outcome: 
- Total counts of chars, digits, and symbols 
- Chars = 8 
- Digits = 3 
- Symbol = 4

```python
str = "P@#yn26at^&i5ve" 
chars = 0
digit = 0
special = 0

for i in str:
    if i.isalpha():
        chars += 1
    elif i.isdigit():
        digit += 1
    else:
        special += 1

print(f"total char is:- {chars}\ntotal digits are:-{digit}\ntotal symbols are :- {special}")
```
